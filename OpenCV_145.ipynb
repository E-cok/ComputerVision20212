## 1.4.5 Morphological Transformations

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 7
img = cv2.imread('j.png', 0)
kernel0 = np.ones((n, n), np.uint8)
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT, (n, n))
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (n, n))
kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS, (n, n))
```

```python
erosion = cv2.erode(img, kernel3, iterations = 1)
dilation = cv2.dilate(img, kernel3, iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel3)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel3)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel3)
```
